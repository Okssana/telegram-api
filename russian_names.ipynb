```python
import pandas as pd

"""
ru_names.csv was manually created from the following sources:
I asked ChatGPT to generate 100 both female and male names and created a csv file from the output.
Currently, the file contains 200 names.
It is in my Google Docs
https://docs.google.com/spreadsheets/d/1Iz6NBs5LkMiLeUqzbOCRHZh6yifyM2JCGRVrFhlo_SE/edit#gid=0
"""
df_ru_names = pd.read_csv('ru_names.csv')       
df_ru_names.head()
```

```python
# Download a csv with the column which contains the found crypto wallets and bank accounts. 
# We can assosiate the wallets/accounts with the names/posts.
# A CSV was created in the chat_Rosich.ipynb
df_chat = pd.read_csv('output/chat_Rosich_cryptoW.csv')
df_chat['text'] = df_chat['text'].fillna('')
# deselect empty rows
df_chat = df_chat[df_chat['text'] != '']
df_chat.head()
```

```python
### FIND MALE NAMES IN TEXT ###

import re
names_list = df_ru_names['ru_male_name'].tolist()

# Function to find names and surnames in text

def find_names(text, names_list):
    found_names = []
    for name in names_list:
        # Regex to find name possibly followed by a surname
        """
        not extracting a surname before a name, improve in the future
        """
        pattern = r'\b' + re.escape(name) + r'\s+[А-Я][а-я]+'
        matches = re.findall(pattern, text)

        found_names.extend(matches)
    return found_names

# Apply the function to each row in df['text']
df_chat['extracted_names'] = df_chat['text'].apply(lambda x: find_names(x, names_list))

# Display the result
display(df_chat.head(10))
```

```python
# split the column 'extracted_names' into rows by comma
df_chat = df_chat.explode('extracted_names')

# drop duplicates in the column 'extracted_names' and 'text'
df_chat = df_chat.drop_duplicates(subset=['extracted_names', 'text'])

# filter out rows with empty values in the column 'extracted_names'
df_chat = df_chat[df_chat['extracted_names'] != '']

# filter out rows with NaN values in the column 'extracted_names'
df_chat = df_chat[df_chat['extracted_names'].notna()]
df_chat.head(5)
```

```python
# save the dataframe to csv
df_chat.to_csv('output/ru_male_names_extracted.csv', index=False, encoding='utf-8-sig')
df_chat
```
